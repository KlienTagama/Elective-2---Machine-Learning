**4. Check Data types**

```python
df.dtypes
```

**5. Dropping irrelevant colums**

```python
df = df.drop(['hotwaterheating', 'airconditioning'], axis=1)
df.head(3)
```

**6. Renaming Columns**

```python
df = df.rename(columns={"bedrooms": "BR", "stories": "STR", "mainroad": "MR"})
df.head(3)
```

**7. Dropping duplicate rows**

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(3)
```

```python
df.count()
```

**8. Dropping missing or null values.**

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

**9. Detecting Outliers using Graphs for Visualization**

**Boxplot**

```python
sns.boxplot(x=df['price'])
```

```python
sns.boxplot(x=df['BR'])
```

```python
sns.boxplot(x=df['area'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

**Scatterplot**

```python
sns.scatterplot(x='price', y='area',
				palette='class', data=df, )
```
